```python
from numpy import *
from matplotlib.pyplot import *
```

```python
n=100
p=0.5
```

```python
def initialize():
    global config, nextconfig
    config = zeros([n,n])
    for x in range(n):
        for y in range(n):
            config[x,y]=1 if random.rand() < p else 0
    nextconfig = zeros([n,n])
```

```python
initialize()
```

```python
nextconfig
```

```python
def update():
    global config, nextconfig
    for x in range(n):
        for y in range(n):
            count = 0
            for dx in [-1,0,1]:
                for dy in [-1,0,1]:
                    count+=config[(x+dx)%n,(y+dy)%n]
                    nextconfig[x,y]=1 if count >=4 else 0
    config,nextconfig=nextconfig,config
                
```

```python
update()
```

```python
config
```

```python
def observe():
    global config,nextconfig
    figure(figsize=[20,20])
    cla()
    imshow(config,cmap='binary')#,figsize=(20,20))
```

```python
p=0.3
initialize()
#figure(figsize=[20,20])
observe()
for i in range(20):
    update()
    observe()
    show()
```
