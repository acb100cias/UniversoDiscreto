```python
def f(r,x):
    return r*x*(1-x)
```

```python
def f2(r,x):
    return f(r,f(r,x))
def f3(r,x):
    return f(r,f(r,f(r,x)))
```

```python
T=arange(0,1.01,0.01)
r=4
figure(figsize=(20,20))
plot(T,f(r,T),label='f')
plot(T,f2(r,T),label=r'$f^2$')
plot(T,f3(r,T),label=r'$f^3$')
plot(T,T)
legend()
```

```python
r=3.9
plot(T,f2(r,T),label='r=%1f'%r)
plot(T,T,label='y=x')
xc=0.73
for i in range(10):
    plot([xc,xc],[xc,f2(r,xc)], color='k', linestyle='--')
    plot([f2(r,xc),xc],[f2(r,xc),f2(r,xc)], color='k', linestyle='--')
    xc=f2(r,xc)
legend()
```

```python
r=3.9
plot(T,f3(r,T),label='r=%1f'%r)
plot(T,T,label='y=x')
xc=0.5
for i in range(10):
    plot([xc,xc],[xc,f3(r,xc)], color='k', linestyle='--')
    plot([f3(r,xc),xc],[f3(r,xc),f3(r,xc)], color='k', linestyle='--')
    xc=f3(r,xc)
legend()
```

```python
def plot_asymptotic_states():
    initialize()
    for t in range(100): # first 100 steps are discarded
        update()
    for t in range(100): # second 100 steps are collected
        update()
        observe()
#figure(figsize=(20,20))
    plot([r] * 100, result, 'b.', alpha = 0.3)
```

```python
figure(figsize=(40,30))
for r in arange(0, 4, 0.001):
    plot_asymptotic_states()

xlabel('r')
ylabel('x')
show()
```
