```python
from numpy import *
from matplotlib.pyplot import *
```

```python
a=1
b=1
c=1
d=1
k=4
```

```python
def X(x,y):
    return array([x+a*x*(1-(x/k))-(1-(1/(b*y+1)))*x,y-c*y+d*x*y])
```

```python
x0=1
y0=1
E=[[x0],[y0]]
```

```python
for i in range(1000):
    x,y=E[0][-1],E[1][-1]
    xt,yt=X(x,y)
    E[0].append(xt)
    E[1].append(yt)
```

```python
plot(E[0],label='presas')
plot(E[1],label='depredadores')
legend()
```

```python
plot(E[0],E[1],'o')
```
