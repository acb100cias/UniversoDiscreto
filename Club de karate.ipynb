```python
gradosCK=[]
for (i,j) in CKZ.degree():
    gradosCK.append(j)
```

```python
grado_prom=array(gradosCK).mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(gradosCK)
plt.axvline(grado_prom,color='r',linestyle='dashed',label='Average Degree')
```

```python
clustering_local=algorithms.cluster.clustering(CKZ)
```

```python
prom_clustering=array(list(clustering_local.values())).mean()
```

```python
plt.hist(list(clustering_local.values()),label=' Clustering Local')
plt.axvline(prom_clustering,color='r',linestyle='dashed',label='clustering local promedio')
plt.legend()
plt.ylabel('Nodos')
plt.title('Clustering')
plt.show()
```
